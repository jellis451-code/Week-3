```
# Team Profile Generator

# Function definition
def print_bio(name, age, goal):
    border = "*" * 25
    print(border)
    print(f"Name: {name}, Age: {age} Goal: {goal}")
    print(border)
    print()  # blank line for separation

# Program logic
print_bio("Jordan", 22, "Learn Python")
print_bio("Ada Lovelace", 36, "Invent Computing")
print_bio("Sherlock Holmes", 40, "Solve Crimes")

```
